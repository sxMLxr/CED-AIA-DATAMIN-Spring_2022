# Revisiting Missing Data... lets see what we have based on our new DataFrame

```python
#some data will contain no data... we should drop those rows
loaddata.dropna()
loaddata.info()
```

```python
#need to drop distance, description and number as they will not fit our predictive model
#labels with text are not useful
loaddata.drop(loaddata[['Distance(mi)','Description','Number']], axis=1, inplace=True)
loaddata.columns
```

```python
loaddata.info()
```

```python
#last lines of the DF can be converted to bool, so well start with Sunrise_Sunset...Astronomical_Twilight [29-32]
loaddata['Sunrise_Sunset']=loaddata['Sunrise_Sunset'].replace(['Day','Night'],[True,False])
loaddata['Civil_Twilight']=loaddata['Civil_Twilight'].replace(['Day','Night'],[True,False])
loaddata['Nautical_Twilight']=loaddata['Nautical_Twilight'].replace(['Day','Night'],[True,False])
loaddata['Astronomical_Twilight']=loaddata['Astronomical_Twilight'].replace(['Day','Night'],[True,False])
```

```python
loaddata.info()
```

```python
#lets see if out bool, statements are successful
print(loaddata['Sunrise_Sunset'][:20].value_counts())
print(loaddata['Civil_Twilight'][:20].value_counts())
print(loaddata['Nautical_Twilight'][:20].value_counts())
print(loaddata['Astronomical_Twilight'][:20].value_counts())
```

```python
#looks good.. now to covert the columns from object type to bool
loaddata['Sunrise_Sunset'] = loaddata['Sunrise_Sunset'].astype('bool')
loaddata['Civil_Twilight'] = loaddata['Civil_Twilight'].astype('bool')
loaddata['Nautical_Twilight'] = loaddata['Nautical_Twilight'].astype('bool')
```

```python
loaddata.info()
```

```python
#Missed one... 
loaddata['Astronomical_Twilight'] = loaddata['Astronomical_Twilight'].astype('bool')
```
